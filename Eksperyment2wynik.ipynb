```python
import sys
sys.path.append('C:\\Users\\korni\\seminarium\\semi\\Lib\\site-packages')
import sklearn
import numpy as np
from skimage.io import imread
import pandas as pd
import os
from skimage.transform import resize
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras import datasets, layers, models
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
dest_path = "C:\\Users\\korni\\seminarium\\semi\\databistest"
data_X = []
data_Y = []
y = 0
list = ['latino', 'rock']

for folder in os.listdir(dest_path):
# for folder in list:
    csv_path = os.path.join(dest_path, folder)
    for file in os.listdir(csv_path):
        data = pd.read_csv(os.path.join(csv_path, file))
        if data.shape == (127, 20):
            data_X.append(data)
            data_Y.append(folder)
data_XP = np.array(data_X)
```

```python
label_encoder = LabelEncoder()
encoded_Y = label_encoder.fit_transform(data_Y)
one_hot_Y = to_categorical(encoded_Y)
```

```python
new_data_x = data_XP.reshape(14394,20*127)
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_data_x, data_Y, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(127, 20, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(len(label_encoder.classes_), activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

```

```python
model.fit(data_X, one_hot_Y, epochs=10, batch_size=32, validation_split=0.2)
```
